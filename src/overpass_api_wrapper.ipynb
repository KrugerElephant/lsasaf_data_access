```python
import overpy

api = overpy.Overpass()

ROME_BOUNDS = (12.531124,41.86426,12.539509,41.87343)  # min_lon, min_lat, max_lon, max_lat

tag = "surface"

query = f'way(41.86426,12.531124,41.86843,12.535509) [{tag}]; (._;>;); out body;'
# fetch all ways and nodes
result = api.query(query)

print(result.ways)


for way in result.ways:
    print("Name: %s" % way.tags.get("name", "n/a"))
    print(f'Feature: {tag.split("=")[0]}, {way.tags.get(tag.split("=")[0], "n/a")}')
    print("  Nodes:")
    for node in way.nodes:
        print("    Lat: %f, Lon: %f" % (node.lat, node.lon))
```
